```python
#Setting up the Data
data = df
x_column = 'Region'
y_column = ' Estimated Unemployment Rate (%)'

#Create the barplot
plt.figure(figsize=(12, 5))
sns.barplot(data=data, x=x_column, y=y_column)

#Rotate X-axis labels vertically
plt.xticks(rotation=90)

#Show the plot
plt.show()
```

```python
#Setting up the data
data = df
x_column = ' Estimated Unemployment Rate (%)'
y_column = ' Estimated Labour Participation Rate (%)'

#Customize colors and Style and Size
plt.figure(figsize=(12, 8))
sns.scatterplot(data= data, x=x_column, y=y_column, color='blue', marker='o', alpha=0.7)

#Adding titles and labels
plt.title("Unemployment Rate vs. Labour Participation Rate")
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Estimated Labour Participation Rate (%)")

#Adding Gridlines for Increased Readability
sns.set(style="whitegrid")

#Show the plot
plt.show()
```

```python
#Setting up the data
data = df
x_column = ' Estimated Unemployment Rate (%)'
y_column = ' Estimated Labour Participation Rate (%)'

#Create a boxplot with custom colors and size
plt.figure(figsize=(12, 6))
sns.boxplot(data= data, x=x_column, y=y_column, palette='Set3')

#Adding titles and labels to the graph
plt.title("Boxplot: Unemployment Rate vs. Labour Participation Rate")
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Estimated Labour Participation Rate (%)")

#Show the plot
plt.show()
```

```python
#Setting up the data
x_column = 'Region'
y_column = ' Estimated Unemployment Rate (%)'
title = 'Unemployment Rates by Region'

#Create the bar chart
fig = px.bar(df, x= x_column, y= y_column, color = 'Region', title= title)

#Show the plot
fig.show()
```

```python
#Setting up the data
x_column = 'Region'
y_column = ' Estimated Unemployment Rate (%)'
title = 'Unemployment Rates by Region'

#Create the box plot
fig = px.box(df, x=x_column, y=y_column, color=x_column, title=title)

#Show the plot
fig.show()
```

```python
#Setting up the data
data = df
x_column = 'Region'
y_column = ' Estimated Unemployment Rate (%)'
title = 'Average Unemployment Rates by Region'

#Creating the histogram using Mean
fig = px.histogram(
    df.groupby(x_column, as_index=False).agg({y_column: 'mean'}),
    x=x_column,
    y=y_column,
    color=x_column,
    title=title
)

#Customize the Y-axis label
fig.update_layout(
    yaxis_title="Average Unemployment Rate (%)"
)

#Show the plot
fig.show()
```

From the Above Analysis:
1. Tripura have the highest average unemployment rate which is 28.35% approximately.
2. It is followed by Haryana with 26.28% and Jharkhand with 20.58% average unemployment.
3. The overall unemployment rate is higher where the labour participation rate is between 30% - 50%.
